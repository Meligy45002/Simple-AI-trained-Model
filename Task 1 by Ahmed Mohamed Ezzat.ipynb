```python
import numpy as np
x = np.arange(1,1001)
y = x + 2
print(x)
print(y)
import tensorflow as tf # call tensorflow library
model =tf.keras.Sequential([                # Sequential is a Keras class for creating models where layers are stacked in a linear sequence,
tf.keras.layers.Dense(1,input_shape=(1,),
activation = 'linear'), # input layer: specifies the input size and intializes the network, Dense(1):means this layers has 1 neuron and it outputs a single value
tf.keras.layers.Dense(10), # hidden layer with 10 neurons : introduce more neurons to learn more complex relationships
tf.keras.layers.Dense(10), # hidden layer with 10 neurons :Adds another fully connected layer with 10 neurons,this increases the model capacity to learn more complex relationships
tf.keras.layers.Dense(1)  # Output layer: outputs the final result
])

model.compile(optimizer = 'adam', loss = 'mse') #using Adam optimizer to adjust model weights
#using MSE(Mean Squared Error) as the metric to evaluate how well the model is performing during training(relative error)

model.fit(x,y, epochs = 200) #fit method trains the model on the given data (x,y) for a specified number of epochs
# x-> input data , y-> target labels(output data) , epoch: One complete pass through the entire training dataset

```

```python
model.predict(np.array([500])) # NumPy array is used to format the input data for TensorFlow models.
```
